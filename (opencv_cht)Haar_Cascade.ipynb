```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow

print(cv2.__version__)
```

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')#正臉
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')#訓練好的人臉－眼睛模型

img = cv2.imread('lenacolor.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)#轉灰階
```

```python
faces = face_cascade.detectMultiScale(gray)#face_cascade分類器   detectMultiScale多臉偵測

for (x,y,w,h) in faces: #4個質
    img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2) #rectangle畫方框

cv2_imshow(img)
```

##多臉偵測

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')#正臉
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')#訓練好的人臉－眼睛模型

img = cv2.imread('family.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)#轉灰階
```

```python
faces = face_cascade.detectMultiScale(gray)#face_cascade分類器   detectMultiScale多臉偵測

for (x,y,w,h) in faces: #4個質
    img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2) #rectangle畫方框

cv2_imshow(img)
```
