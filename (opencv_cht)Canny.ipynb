# 使用不同的threshold值來看`lena.bmp`的檢測結果

```
o=cv2.imread("lena.bmp",cv2.IMREAD_GRAYSCALE)
r1=cv2.Canny(o,128,200)#第一個設定值
r2=cv2.Canny(o,32,128)#第二個設定值
cv2_imshow(o)
cv2_imshow(r1)
cv2_imshow(r2)
```

```
o=cv2.imread("penguin.jpg",cv2.IMREAD_GRAYSCALE)
r1=cv2.Canny(o,128,200)
r2=cv2.Canny(o,32,128)
cv2_imshow(o)
cv2_imshow(r1)
cv2_imshow(r2)
```
